## You can choose barchart for discrete data sets

1. In this example we are plotting number of medals won by a person on differnt games

```python
from matplotlib import pyplot as plt

# we can plot charts inline with following:
%matplotlib inline  
```

```python
games = ["Game 1", "Game 2", "Game 3", "Game 4", "Game 5"]
medals = [3, 7, 2, 12, 13]

ind = list(range(len(games))) # x-coordiantes
# # plot bars with left x-coordinates [ind], heights [medals]
plt.bar(ind, medals,align='center', alpha=0.5)
plt.ylabel("# of Medals")
plt.title("Medals won in different Games")
# label x-axis with game names at bar centers
plt.xticks(ind, games)
plt.show()
```

```python
plt.barh(ind, medals, align='center', alpha=0.5)
plt.yticks(ind, games)
plt.xlabel('# of Medals')
plt.title('Medals won in different Games')
 
plt.show()
```
